```python
#special typing effect program
import time,os,sys
import pandas as pd
import getpass

def typingprint(text):
    for character in text:
        sys.stdout.write(character)
        sys.stdout.flush()
        time.sleep(0.0005)


def typinginput(text):
    for character in text:
        sys.stdout.write(character)
        sys.stdout.flush()
        time.sleep(0.05)
    value = input()
    return value


def clearScreen():
    os.system("clear")


# actual game
# display screen
typingprint("HELLO BOTH OF YOU! \n")
time.sleep(1)
typingprint("WELCOME TO ODD EVEN GAME !\n")
time.sleep(1)
typingprint("RULES:\n 1) Each player will start with 10 marbles\n")
typingprint(
    "2) One player is tasked with picking some number of marbles, the opposite player has to guess if the number of marbles picked is odd or even.\n")
typingprint(
    "If they're correct, the player must give those marbles to the opponent, and if they're wrong, the opponent must give the player as many marbles that he had picked\n")
typingprint("3) The game will be played turn by turn by each player\n")
typingprint("4) The player to have all 20 marbles wins the game!\n")
time.sleep(1)

choice = typinginput("would you like to play?\n YES-[y] \n NO-[n]\n")
if (choice == 'n'):
    typingprint("Good bye!..\n")
    time.sleep(1)
    typingprint("See you soon, closing in 3..")
    time.sleep(1)
    typingprint("2..")
    time.sleep(1)
    typingprint("1..")
    time.sleep(1)
    clearScreen()
else:
    typingprint("Lets start the game...\n")
    time.sleep(1)
    typingprint("But before that we would like to know your names")
    name1 = typinginput("\nplayer1 name:")
    name2 = typinginput("\nplayer2 name:")

    # starting of the game
    playerscore1 = 10
    playerscore2 = 10
    list1 = []
    list2 = []
    list1guess = []
    list2guess = []
    list1guess.append(0)
    list2guess.append(0)
    list1.append(playerscore1)
    list2.append(playerscore2)
    playerturn = True
    typingprint("\n" + name1 + " score=10 \n" + name2 + " score=10 \n")


    # class for printing and calaculation of score
    class scorecalculator():

        def __init__(self, playerscore1, playerscore2, playerchoice1, playerchoice2, token1, name1, name2):
            self.ps1 = playerscore1
            self.ps2 = playerscore2
            self.pc1 = playerchoice1
            self.pc2 = playerchoice2
            self.t1 = token1
            self.n1 = name1
            self.n2 = name2

        def calculateplayer1(self):

            if (self.t1) == 1:

                self.ps1 = self.ps1 + self.pc2
                self.ps2 = self.ps2 - self.pc2


            else:

                self.ps2 = self.ps2 + self.pc1
                self.ps1 = self.ps1 - self.pc1

            typingprint("\n" + name1 + "'s marbles left=" + str(self.ps1) + "\n" + name2 + "'s marbles left=" + str(
                self.ps2) + "\n\n")
            return (self.ps1, self.ps2)

        def calculateplayer2(self):

            if (self.t1) == 1:
                self.ps2 = self.ps2 + self.pc1
                self.ps1 = self.ps1 - self.pc1

            else:
                self.ps1 = self.ps1 + self.pc2
                self.ps2 = self.ps2 - self.pc2
            typingprint("\n" + name1 + "'s marbles left=" + str(self.ps1) + "\n" + name2 + "'s marbles left=" + str(
                self.ps2) + "\n\n")
            return (self.ps1, self.ps2)


    # function for comparing
    def Comparingfunction(playerguess, playername):

        if (playerguess == playername):
            t1 = 1
            typingprint("\n YOU GUESSED IT RIGHT! :)\n")
            return t1

        else:
            t1 = 0
            print("\n YOU GUESSED IT WRONG! :[\n")
            return t1


    # main function
    while True:

        while playerturn is True:
            playerchoice1 = 0
            playerchoice2 = 0
            playerchoice1 = int(getpass.getpass(prompt=name1 + " Pick your marbles for opponent to guess:"))

            if (playerchoice1 % 2) == 0:
                playerans = 'E'
            else:
                playerans = 'O'

            playerchoice2 = int(getpass.getpass(prompt=name2 + " Pick your marbles:"))
            playerguess = typinginput("\n ODD-[O] \n EVEN-[E] \n" + name2 + " YOUR CHOICE=")

            # function calling
            token1 = Comparingfunction(playerguess, playerans)

            # class calling
            callingclass = scorecalculator(playerscore1, playerscore2, playerchoice1, playerchoice2, token1, name1,
                                           name2)
            playerscore1, playerscore2 = callingclass.calculateplayer2()

            list1guess.append(playerchoice1)
            list2guess.append(playerchoice2)
            list1.append(playerscore1)
            list2.append(playerscore2)
            playerturn = False

        if (playerscore1 == 0 or playerscore2 == 0):
            break

        while playerturn is False:
            playerchoice1 = 0
            playerchoice2 = 0
            playerchoice2 = int(getpass.getpass(prompt=name2 + " Pick your marbles for opponent to guess:"))

            if (playerchoice2 % 2) == 0:
                playerans = 'E'
            else:
                playerans = 'O'

            playerchoice1 = int(getpass.getpass(prompt=name1 + " Pick your marbles:"))
            playerguess = typinginput("\n ODD-[O] \n EVEN-[E] \n" + name1 + " YOUR CHOICE=")

            # function calling
            token1 = Comparingfunction(playerguess, playerans)

            # class calling
            callingclass = scorecalculator(playerscore1, playerscore2, playerchoice1, playerchoice2, token1, name1,
                                           name2)
            playerscore1, playerscore2 = callingclass.calculateplayer1()

            list1guess.append(playerchoice1)
            list2guess.append(playerchoice2)
            list1.append(playerscore1)
            list2.append(playerscore2)
            playerturn = True

        if (playerscore1 == 0 or playerscore2 == 0):
            break

    # DECLARING WINNER
    if (playerscore1 == 20):
        typingprint("Congrats " + name1 + " you WIN! :)\n")
    else:
        typingprint("Congrats " + name2 + " you WIN! :)\n")

    # table using pandas
    typingprint("GAME ANAYLSIS:\n\n")
    data = {'PLAYER1': list1, '  PLAYER1 GUESSES': list1guess, '  PLAYER2': list2, '  PLAYER2 GUESSES': list2guess}
    analysis = pd.DataFrame(data)
    print(analysis)

    # ENDING
   #typingprint("\n KALPIT PRABHU-60001200080\n CHIRAYU PATEL-60001200085c")
    typingprint("\n Good bye!..\n")
    time.sleep(1)
    typingprint("Thanks for playing, See you soon, closing in 3..")
    time.sleep(1)
    typingprint("2..")
    time.sleep(1)
    typingprint("1..")
    time.sleep(1)
    clearScreen()
```
